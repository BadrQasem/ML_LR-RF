## **First ML Project**

## **Load data**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

## **Datat preperation**

### Data seperation as x and y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

### Data spilitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

## **Models Building**

### Linear Regression

####Traing the model


```python
from sklearn.linear_model import LinearRegression
Ir = LinearRegression()
Ir.fit(x_train, y_train)
```

#### Applying the model to make a prediction



```python
y_Ir_train_pred = Ir.predict(x_train)
y_Ir_test_pred = Ir.predict(x_test)
```

```python
y_Ir_train_pred
```

```python
y_Ir_test_pred
```

#### Evaluate model performance



Here we will compare the ecual value with the trained predicted values

```python
from sklearn.metrics import mean_squared_error, r2_score

Ir_train_mse = mean_squared_error(y_train, y_Ir_train_pred)
Ir_train_r2 = r2_score(y_train, y_Ir_train_pred)

Ir_test_mse = mean_squared_error(y_test, y_Ir_test_pred)
Ir_test_r2 = r2_score(y_test, y_Ir_test_pred)
```

```python
print ('LR MSE (Train) :', Ir_train_mse)
print ('LR R2 (Train) :', Ir_train_r2)
print ('LR MSE (Test) :', Ir_test_mse)
print ('LR R2 (Test) :', Ir_test_r2)
```

```python
Ir_results = pd.DataFrame(['Linear Regression', Ir_train_mse, Ir_train_r2, Ir_test_mse, Ir_test_r2]). transpose()
Ir_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
Ir_results
```

### Random Forest

#### Training the Model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

#### Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

#### Evaluate model performance


```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]). transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

### Model Comparison

```python
df_models = pd.concat([Ir_results, rf_results], axis=0)
df_models.reset_index(drop=True)
```

## Data Visualization of Prediction results

### Visualion of Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_Ir_train_pred, c="#7CAE00", alpha=0.3)

z = np.polyfit(y_train, y_Ir_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
plt.title('Linear Regression: Experimental vs Predicted LogS')
plt.text(min(y_train), max(y_Ir_train_pred), f'MSE: {Ir_train_mse:.2f}\nR^2: {Ir_train_r2:.2f}', fontsize=9)
plt.show()
```

### Visualization both Models

```python
plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_Ir_train_pred, c="#7CAE00", alpha=0.3, label='Linear Regression')
z = np.polyfit(y_train, y_Ir_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train), '#F8766D', label='Linear Regression Fit')

plt.scatter(x=y_train, y=y_rf_train_pred, c="#00BFC4", alpha=0.3, label='Random Forest')
z = np.polyfit(y_train, y_rf_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train), '#619CFF', label='Random Forest Fit')

plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')
plt.legend()
plt.show()
```

### Random Forest Regressor model consists of multiple decision trees

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Assuming 'rf' is your trained RandomForestRegressor model
# Visualize the first tree in the forest
plt.figure(figsize=(25,10))
plot_tree(rf.estimators_[0], feature_names=x_train.columns, filled=True, rounded=True)
plt.show()
```

### visualize the importance of each feature

```python
import matplotlib.pyplot as plt
importances = rf.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(10, 6))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [x_train.columns[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

### Partial Dependence Plots for single features

```python
from sklearn.inspection import PartialDependenceDisplay
import matplotlib.pyplot as plt

# Partial Dependence Plots for single features
features = [0, 1, 2]  # Replace with the indices or names of the features you want to analyze
PartialDependenceDisplay.from_estimator(rf, x_train, features, kind='average')  # or kind='both'
plt.show()

# Partial Dependence Plots for feature pairs
feature_pairs = [(0, 1), (1, 2)]  # Replace with feature pairs
PartialDependenceDisplay.from_estimator(rf, x_train, feature_pairs, kind='average')  # or kind='both'
plt.show()
```

### structure of a specific decision tree


```python
# Assuming 'rf' is your trained RandomForestRegressor model
# Visualize the first tree in the forest
plt.figure(figsize=(25,10))
plot_tree(rf.estimators_[0], feature_names=x_train.columns, filled=True, rounded=True)
plt.show()
```

### the behavior of the whole random forest

```python
from sklearn.tree import DecisionTreeRegressor

# Train a decision tree to approximate the random forest
tree_approx = DecisionTreeRegressor(max_depth=3) # Adjust max_depth as needed
tree_approx.fit(x_train, y_rf_train_pred) # Train on RF predictions


#Visualize the decision tree
plt.figure(figsize=(25,10))
plot_tree(tree_approx, feature_names=x_train.columns, filled=True, rounded=True)
plt.show()
```
